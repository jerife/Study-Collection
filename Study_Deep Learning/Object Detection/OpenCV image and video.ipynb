### 영상처리

```python
!mkdir ./data
!wget -O /content/data/Night_Day_Chase.mp4 https://github.com/chulminkw/DLCV/blob/master/data/video/Night_Day_Chase.mp4?raw=true
```

```python
video_input_path = '/content/data/Night_Day_Chase.mp4'
video_output_path = '/content/data/Night_Day_Chase_out.mp4'

cap = cv2.VideoCapture(video_input_path)
codec = cv2.VideoWriter_fourcc(*'XVID')

vid_size = (round(cap.get(cv2.CAP_PROP_FRAME_WIDTH)),round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
vid_fps = cap.get(cv2.CAP_PROP_FPS)

vid_writer = cv2.VideoWriter(video_output_path, codec, vid_fps, vid_size)
frame_count = cap.get(cv2.CAP_PROP_FRAME_COUNT)
print("Frame 총 개수 : {} | FPS : {} | SIZE : {}".format(frame_count, round(vid_fps), vid_size))
```

```python
green_color = (0, 255, 0)
red_color = (0, 0, 255)

start = time.time()
index = 0

while True:
  hasFrame, img_frame = cap.read()
  if not hasFrame:
    print("더 이상 처리할 Frame이 없습니다.")
    break
  index += 1
  print("Frame : {} 처리 완료".format(index))

  cv2.rectangle(img_frame, (300,100,800,400), green_color, thickness=2)
  text = "Frame : {}".format(index)
  cv2.putText(img_frame, text, (300,95), cv2.FONT_HERSHEY_SIMPLEX, 0.7, red_color, 1) 

  vid_writer.write(img_frame)

print("Write 완료 | 소요 시간: {}".format(round(time.time()-start)))
vid_writer.release()
cap.release()   
```
