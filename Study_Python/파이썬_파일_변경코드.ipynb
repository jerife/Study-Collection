# 파일 명 수정 코드

```python
import os
from glob import glob
```

```python
# 주어진 디렉토리에 있는 항목들의 이름을 담고 있는 리스트를 반환합니다.
# 리스트는 임의의 순서대로 나열됩니다.
file_path = '/Users/jerife/Desktop/sweatshirt_dataset'
file_names = os.listdir(file_path)
file_names
```

```python
i = 1
for name in file_names:
    src = os.path.join(file_path, name)
    dst = "3-" +str(i) + '.jpg'
    dst = os.path.join(file_path, dst)
    os.rename(src, dst)
    i += 1
```

# Text 파일 내용 수정코드

```python
file_paths = glob('/Users/jerife/Desktop/sweatshirt_dataset/*.txt')
file_paths
```

```python
for file_path in file_paths[:2]:
    with open(file_path,'r') as f:
        lines = f.readlines()
        print(lines)
        
    with open(file_path,'w') as f:
        f.write('2' + lines[0][2:])
        
        
    with open(file_path,'r') as f:
        lines = f.readlines()
        print(lines,'\n\n\n')
```

```python
for file_path in file_paths[2:]:
    with open(file_path,'r') as f:
        lines = f.readlines()
        print(lines)
```
