#### GridSearchCV

```python
params ={
    'max_depth' :[1,2,3],
    'min_samples_split':[2,3]
}
```

```python
iris_data = iris.data
iris_label = iris.target
x_train,x_test,y_train,y_test = train_test_split(iris_data, iris_label, test_size = 0.2)
dt_clf=DecisionTreeClassifier()
```

```python
grid = GridSearchCV(dt_clf, param_grid = params, cv=3)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_estimator_, grid.best_params_, grid.best_score_
```

```python
estimator = grid.best_estimator_
pred = estimator.predict(x_test)
print(accuracy_score(pred,y_test))
```

```python
def cate_age(age):
    cat=''
    if age <= -1: cat = 'Uknown'
    elif age<=5: cat = 'Baby'
    elif age<=12: cat = 'Child'
    elif age<=18: cat = 'Teenager'
    elif age<=25: cat = 'Student'
    elif age<=35: cat = 'Adult'
    elif age<=60: cat = 'Elderly'
        
    return cat    
```

```python
titanic_df['Age_categori'] = titanic_df['Age'].apply(lambda x: cate_age(x))
titanic_df
```

```python
titanic_df.groupby('Age_categori')['Survived'].value_counts().sort_index()
```
