```python
def age(age):
    if age > 60:
        re = 'Older'
    elif age > 20:
        re = 'Adult'
    else:
        re = 'Child'
    return re
```

```python
titanic_df['Who'] = titanic_df['Age'].apply(lambda x : age(x))
titanic_df
```
